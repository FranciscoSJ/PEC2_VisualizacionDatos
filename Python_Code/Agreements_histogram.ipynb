```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("Datasets/Agreements.csv")
data.head()
```

```python
#Año
data["Year"]=[data["Dat"][i][0:4] for i in range(len(data["Dat"]))]
year_plot = sns.catplot(x="Year", kind="count", palette="brg", data=data, size=5)
year_plot.set_axis_labels( "Año", "Frecuencia")
year_plot.fig.set_size_inches(16,8)
```
